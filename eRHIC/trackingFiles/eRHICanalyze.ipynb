# Analyze results from eRHIC electron ring tracking

We track an electron for several damping times through eRHIC and plot the spin evolution

## observe orbit and spin evolution from .fai file

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import os
import math

FL=os.getcwd()
print FL
dm1val,xval,yval,sx,sy,sz,IPASS=np.loadtxt('%s/%s'%(FL,'eRHIC.fai'),skiprows=4,usecols=(8,9,11,19,20,21,37),unpack=True)
xvalmeter=xval/100
yvalmeter=yval/100

print(np.size(IPASS))
print(np.size(xval))

numpart=1
numturns=234801
nvals=np.size(IPASS)

nplot=1000

plt.figure(figsize=(5,2))
plt.plot(IPASS[0:nplot],xval[0:nplot],label=r'$x$')
plt.legend()
plt.show()

plt.figure(figsize=(5,2))
plt.plot(IPASS[0:nplot],yval[0:nplot],label=r'$y$')
plt.legend()
plt.show()

plt.figure(figsize=(5,2))
plt.plot(IPASS[0:nplot],dm1val[0:nplot],label=r'$\frac{dp}{p}$')
plt.legend()
plt.show()


plt.figure(figsize=(5,2))
plt.plot(IPASS[0:nplot],sx[0:nplot],label=r'$s_x$')
plt.legend()
plt.show()

plt.figure(figsize=(5,2))
plt.plot(IPASS[0:nplot],sy[0:nplot],label=r'$s_y$')
plt.legend()
plt.show()

plt.figure(figsize=(5,2))
plt.plot(IPASS[0:nplot],sz[0:nplot],label=r'$s_z$')
plt.legend()
plt.show()
```
