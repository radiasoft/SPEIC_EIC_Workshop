```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import os
FL=os.getcwd()
print FL

S,sx,sy,sz,mods,npart=np.loadtxt('%s/%s'%(FL,'jleic_noerr.fai'),skiprows=3,usecols=(13,19,20,21,22,26),unpack=True)
yco,zco=np.loadtxt('%s/%s'%(FL,'jleic_noerr.fai'),skiprows=3,usecols=(9,11),unpack=True)
sx0=sx[0::3]  #comments...
sxd=sx[1::3]
sxmd=sx[2::3]

sy0=sy[0::3]
syd=sy[1::3]
symd=sy[2::3]

sz0=sz[0::3]
szd=sz[1::3]
szmd=sz[2::3]

yco0=yco[0::3]
ycod=yco[1::3]
ycomd=yco[2::3]

zco0=zco[0::3]
zcod=zco[1::3]
zcomd=zco[2::3]

delta=1e-4

dnx=(sxd-sxmd)
dny=(syd-symd)
dnz=(szd-szmd)

dnddAbs=np.sqrt(dnx**2+dny**2+dnz**2)/(2*delta)

Sm=S/100  # S coordinate in meters
Sm0=Sm[::3]
```

```python
plt.figure(figsize=(5,5))
#plt.plot(Sm0,dnx,label=r'$Longitudinal d=0$')
#plt.plot(Sm0,dny,label=r'$Longitudinal d>0$')
#plt.plot(Sm0,dnz,label=r'$Longitudinal d<0$')
plt.plot(Sm0,dnddAbs,label=r'$\frac{d\hat n}{d\delta}$')


#plt.ylim([40,50])
#plt.xlim([0,np.max(Sm)])
plt.legend()
plt.ylabel(r'$Spin Components $')
plt.xlabel(r'$\rm s [m]$')
plt.tight_layout()
plt.savefig('jleic_noerr.png')
plt.show()
```
